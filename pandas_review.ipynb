#### Adding/Removing

```python
# df['new_col_name'] = some value
import numpy as np
df_new = df.copy()
df_new['price'] = 4.99
df_new['new_price'] = np.where(df_new['Coffee Type'] == 'Espresso', 3.99, 5.99)
df_new.head()
```

```python
# df.drop('0') # drop row 0
df_new.drop(columns=["price"], 
        # inplace=True, # if you really want to modify the df
        )
```

```python
df['revenue'] = df['Units Sold'] * df['new_price']
df.rename(columns={'Units Sold': 'Units_Sold',
                   'Coffee Type': 'Coffee_Type'}, inplace=True)

df.head()
```

```python
bios_new = bios.copy()
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]

bios_new.query('first_name == "Keith"')
```

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'],
                                           # errors='coerce,format="%Y-%m-%d" # https://strftime.org/
                                           ) 
bios_new.info()
```

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new[['name', 'born_year']]
```

```python
bios_new['height_category'] = bios_new['height_cm'].apply(lambda x: 'Short' if x<165 else 
                                                          ('Average' if x<185 else 'Tall'))
bios_new.head()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return "LightWeight"
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return "MiddleWeight"
    else:
        return "HeavyWeight"
    
bios_new['category'] = bios_new.apply(categorize_athlete, axis=1) # 1 for veritical, 0 for horizontal
bios_new.head()
```
